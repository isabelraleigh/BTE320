```python
sales = float(input("Enter sales value ($): "))

cogs = .4 * sales
gross_profit = sales - cogs

print(f"Sales: ${sales}")
print(f"COGS: ${cogs:}")
print(f"Gross Profit: ${gross_profit}")
```

```python
sales = float(input("Enter sales value ($): "))
tax_rate = float(input("Enter tax rate (%): "))

cogs = 0.4 * sales
gross_profit = sales - cogs
taxes = gross_profit * tax_rate
net_income = gross_profit - taxes

print(f"Sales: ${sales}")
print(f"COGS: ${cogs:}")
print(f"Gross Profit: ${gross_profit}")
print(f"Taxes: ${taxes}")
print(f"Net Profit: ${net_income}")
```

```python
num1 = float(input("Enter number: "))
num2 = float(input("Enter number: "))
oper = input("Enter operator (+, -, *, /): ")

if oper == "+":
  print(num1 + num2)
if oper == "-":
  print(num1 - num2)
if oper == "*":
  print(num1 * num2)
if oper == "/":
  print(num1/num2)
```
