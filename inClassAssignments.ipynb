```python
sales = float(input("Enter sales value ($): "))
tax_rate = float(input("Enter tax rate (%): "))

cogs = 0.4 * sales
gross_profit = sales - cogs
taxes = gross_profit * tax_rate/100
net_income = gross_profit - taxes

print(f"Sales: ${sales}")
print(f"COGS: ${cogs:}")
print(f"Gross Profit: ${gross_profit}")
print(f"Taxes: ${taxes}")
print(f"Net Profit: ${net_income}")
```

```python
num1 = float(input("Enter number: "))
num2 = float(input("Enter number: "))
oper = input("Enter operator (+, -, *, /, etc.): ")

if oper == "+":
  res = num1 + num2
elif oper == "-":
  res = num1 - num2
elif oper == "*":
  res = num1 * num2
elif oper == "/":
  res = num1 / num2
elif oper == "**":
  res = num1 ** num2
elif oper == "//":
  res = num1 // num2
elif oper == "%":
  res = num1 % num2
else:
  print("Invalid operator")
  res = None


print(f"Result: {res}")
```

```python
print(" Hour Fee")
for hour in range(1,9,1):
  fee = 5 + 2.5 * hour
  if fee < 10:
    fee = 10.0
  elif fee > 25:
    fee = 20.0
  print(f" {hour}: " , fee)

```

```python
import math

ticketprice = 10
defaultAudience = 20

print(" Ad$ Profit")
for adMoney in range(0,201,25):
  revenue =  (defaultAudience + 2 * round( math.sqrt(adMoney))) * ticketprice
  costs = 200 + adMoney
  profit = revenue - costs
  print(f"{float(adMoney)}  {float(profit)}")

```

```python
print(" Hour Fee")
hour = 1
while hour < 9: # or while hour <= 8
  fee = 5 + 2.5 * hour
  if fee < 10:
    fee = 10.0
  elif fee > 25:
    fee = 20.0
  hour += 1
  print(f" {hour}: " , fee)

```

```python
import math

ticketprice = 10
defaultAudience = 20
adMoney = 0
print(" Ad$ Profit")
while adMoney < 201:
  revenue =  (defaultAudience + 2 * round( math.sqrt(adMoney))) * ticketprice
  costs = 200 + adMoney
  profit = revenue - costs
  print(f"{float(adMoney)}  {float(profit)}")
  adMoney += 25
```

```python
import random
print("Welcome to rock, paper, scissors!")
play = input("Do you want to play? (y/n)")

while play == "y":
  userChoice = input("Enter your choice (rock, paper, scissors): ")
  computerChoice = random.choice(['rock', 'paper', 'scissors'])
  print(f"Computer chose" ,{computerChoice})
  if userChoice == computerChoice:
    print("It's a tie")
  elif (userChoice == "rock" and computerChoice == "scissors") or  (userChoice == "paper" and computerChoice == "rock") or  (userChoice == "scissors" and computerChoice == "paper"):
    print("You win")
  else:
    print("You lose")
  play = input("Do you want to play? (y/n)")




```
